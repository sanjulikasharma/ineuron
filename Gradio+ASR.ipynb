```python
import gradio as gr
import torchaudio
import torch
from transformers import pipeline
```

```python
asr_model = pipeline("automatic-speech-recognition", model="openai/whisper-large")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def transcribe(audio_path):
    # Receive audio input
    # print(audio_path)
    audio_input, sample_rate = torchaudio.load(audio_path)

    # Pre-process the audio data
    # For simplicity, we'll use torchaudio's resampling to 16kHz
    resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
    preprocessed_audio = resampler(audio_input)

    # Pass the data through the ASR model
    model_output = asr_model(preprocessed_audio.numpy()[0])

    # Post-process the model output
    # The ASR model returns a list of hypotheses, we'll take the first one
    text_output = model_output['text']

    return text_output
# Define the Gradio interface
interface = gr.Interface(
    title='Real-time AI-based Audio Transcription Web App',
    fn=transcribe,
    inputs=[
        gr.components.Audio(sources=["upload", "microphone"], type="filepath", label="Upload/Record Audio File"),
    ],
    outputs=[
        gr.components.Textbox()
    ],
    live=True
)

# Launch the interface
interface.launch(debug=True)
```
